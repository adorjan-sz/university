Egy hatékony módja a képek betöltésének: a .tar fájl betöltése a memóriában egyben és kicsomagolás ott.

```python
color_ims = []
label_ims = []

# unpacking tar content directly into memory and decoding jpg/png images from there to avoid working with many small files on disk
with tarfile.open("PascalVOC_annbsc_tiny.tar", "r") as tar:
    tar_members = tar.getmembers()
    tar_jpg_members = [member for member in tar_members if member.name[-4:] == '.jpg']  # list only jpg images, not label images
    for jpg_member in tar_jpg_members:
        filename_noext = os.path.split(jpg_member.name)[-1][:-4]

        # get next jpg image
        file_in_tar = tar.extractfile(jpg_member)
        byte_arr = np.asarray(bytearray(file_in_tar.read()), dtype=np.uint8)
        pil_image = Image.open(io.BytesIO(byte_arr))
        npy_image = np.asarray(pil_image)
        assert npy_image.dtype == np.uint8
        color_ims.append(npy_image)

        # get label image corresponding to the jpg image
        label_member_name = 'PascalVOC_annbsc_data/SegmentationClass/' + filename_noext + '.png'
        file_in_tar = tar.extractfile(label_member_name)
        byte_arr = np.asarray(bytearray(file_in_tar.read()), dtype=np.uint8)
        pil_image = Image.open(io.BytesIO(byte_arr))        # using PIL instead of cv2 as it supports loading palette coded png images as single channel integer arrays
        npy_image = np.asarray(pil_image)
        assert npy_image.dtype == np.uint8
        label_ims.append(npy_image)

color_ims = np.stack(color_ims, axis=0)
label_ims = np.stack(label_ims, axis=0)

print("Color images shape, dtype:", color_ims.shape, color_ims.dtype)   # axis order: NHWC (n_images, height, width, n_channels)
print("Label images (masks) shape, dtype:", label_ims.shape, label_ims.dtype)
```

## Az adatok elkészítése a betanításhoz

A (multi-class) klasszifikációs feladathoz a kategorikus címkék előállítása:
- Ha autó (car) van a képen $\to$ 0
- Egyébként, ha macska (cat) van a képen $\to$ 1
- Egyébként, ha kutya (dog) van a képen $\to$ 2
- Egyébként eldobjuk a képet.

```python
car_id = ORIG_CATEGORY_NAMES.index('car')   # == 7
cat_id = ORIG_CATEGORY_NAMES.index('cat')   # == 8
dog_id = ORIG_CATEGORY_NAMES.index('dog')   # == 12
```

```python
RUN_SOLUTION1 = False

if RUN_SOLUTION1 is True:
    # SOLUTION #1: Create categorical labels for the selected images. Partly vectorized, handling categories separately.

    n_images = label_ims.shape[0]

    car_present_mask = np.any(label_ims == car_id, axis=(1,2))     # (n_images,h,w) -> (n_images,) bool_
    cat_present_mask = np.any(label_ims == cat_id, axis=(1,2))     # (n_images,h,w) -> (n_images,) bool_
    dog_present_mask = np.any(label_ims == dog_id, axis=(1,2))     # (n_images,h,w) -> (n_images,) bool_

    cl_labels = np.full((n_images,), dtype=np.int64, fill_value=99)          # we indicate invalid value with 99 now
    cl_labels[dog_present_mask] = 2                                   #    cl_labels:
    cl_labels[cat_present_mask] = 1                                   #        if car present in image -> label = 0
    cl_labels[car_present_mask] = 0                                   #        elif cat present in image -> label = 1
                                                                    #        elif dog present in image -> label = 2
                                                                    #        (a few images contain more than one of these categories but we are not bothered by that for now - this will add a small amount of noise to the data)

    kept_images_mask = cl_labels != 99               # (n_images,) bool_

    color_ims = color_ims[kept_images_mask, :, :, :]         # (n_kept_images, h, w, n_channels=3) uint8
    cl_labels = cl_labels[kept_images_mask]                  # (n_kept_images,) int64

else:

    # SOLUTION #2: Create categorical labels for the selected images. Fully vectorized.

    car_cat_dog_present = np.any(label_ims[:,:,:,None] == [car_id, cat_id, dog_id], axis=(1,2))    # (n,h,w,1) == (3,) -> (n,h,w,3); any: (n,h,w,3) -> (n,3)
    car_or_cat_or_dog_present = np.any(car_cat_dog_present, axis=1)   # (n_images, 3) -> (n_images,) bool_

    color_ims = color_ims[car_or_cat_or_dog_present, :, :, :]         # (n_kept_images, h, w, n_channels=3)
    n_kept_images = color_ims.shape[0]
    car_cat_dog_present_in_kept_imgs = car_cat_dog_present[car_or_cat_or_dog_present, :]   # (n_kept_images, 3:[car, cat, dog])

    cl_labels = np.argmax(car_cat_dog_present_in_kept_imgs, axis=1)   # (n_kept_images,)
                                                                    #    cl_labels:
                                                                    #        if car present in image -> label = 0
                                                                    #        elif cat present in image -> label = 1
                                                                    #        elif dog present in image -> label = 2
                                                                    #        (no else, as at least one category was present)
                                                                    #        (a few images contain more than one of these categories but we are not bothered by that for now - this will add a small amount of noise to the data)


print("Labels distribution: ", np.bincount(cl_labels))
```

Train/val/test split.

```python
# Here, the samples can be optionally randomly shuffled
#    Note, that images and labels must be shuffled in the same order (e.g., generating a random permutation of arange and indexing with that into both arrays)


# dataset split through basic indexing
#    alternative: scikit-learn - sklearn.model_selection.train_test_split
TRAINING_SET_RATIO = 0.5
VALIDATION_SET_RATIO = 0.25
assert TRAINING_SET_RATIO + VALIDATION_SET_RATIO < 1

n_training_samples = int(color_ims.shape[0] * TRAINING_SET_RATIO)
n_validation_samples = int(color_ims.shape[0] * VALIDATION_SET_RATIO)
validation_offset = n_training_samples
test_offset = n_training_samples + n_validation_samples

color_ims_train = color_ims[:validation_offset, :, :, :]
cl_labels_train = cl_labels[:validation_offset]
color_ims_val = color_ims[validation_offset:test_offset, :, :, :]
cl_labels_val = cl_labels[validation_offset:test_offset]
color_ims_test = color_ims[test_offset:, :, :, :]
cl_labels_test = cl_labels[test_offset:]

print(f"Training input images shape: {color_ims_train.shape}, classification labels shape: {cl_labels_train.shape}")
print(f"Validation input image shape: {color_ims_val.shape}, classification labels shape: {cl_labels_val.shape}")
print(f"Test input images shape: {color_ims_test.shape}, classification labels shape: {cl_labels_test.shape}")
```

Az adat-iterátor elkészítése (`torch.utils.data.Dataset` és `torch.utils.data.DataLoader` segítségével, egyelőre `torchvision.transforms` nélkül).

```python
# very simple iterator: only basic type/dtype conversions happen inside

class ImageClassificationDataset(Dataset):
    def __init__(self, color_ims_split, cl_labels_split):
        # expecting: color_ims_split (n_img, 240, 320, 3), cl_labels_split (n_img,)
        assert color_ims_split.shape[0] == cl_labels_split.shape[0]
        assert color_ims_split.shape[1:] == (240, 320, 3)

        color_ims_split = color_ims_split.transpose(0, 3, 1, 2)   # numpy transpose, swap axis order from NHWC to NCHW: ndarray(n_img, 240, 320, 3) -> ndarray(n_img, 3, 240, 320)
                                                                  #     same op in torch: torch.permute()
                                                                  # np.swapaxes() is similar to torch.transpose()
        self.xs_split_t = torch.from_numpy(color_ims_split)       # tensor(n_img, 3, 240, 320) uint8; keep it in uint8 as it takes up much less space compared to float32
        self.ys_split_t = torch.from_numpy(cl_labels_split)       # tensor(n_img,) int64; multi-class labels

    def __len__(self):
        return self.xs_split_t.shape[0]

    def __getitem__(self, idx):
        x = self.xs_split_t[idx,:,:,:].to(torch.float32)    # convert the requested sample to float32 only on demand to save up memory (see comments above)
        y = self.ys_split_t[idx]
        return x, y                # x: tensor(3, h, w) fl32 in range [0, 255];    y: scalar int64
                                   #     -> DataLoader will produce (tensor(batch_size, 3, h, w) fl32, tensor(batch_size,) int64)


train_dataset = ImageClassificationDataset(color_ims_train, cl_labels_train)
val_dataset = ImageClassificationDataset(color_ims_val, cl_labels_val)
test_dataset = ImageClassificationDataset(color_ims_test, cl_labels_test)

dataloader_cl_train = DataLoader(train_dataset, batch_size=32, shuffle=True)
dataloader_cl_val = DataLoader(val_dataset, batch_size=32, shuffle=True)
dataloader_cl_test = DataLoader(test_dataset, batch_size=32, shuffle=False)  # not shuffling!
```

## Az egyszerű konvolúcós háló betanítása képek (multi-class) klasszifikációjára. **Csak CPU.**

Egyszerű konvolúciós háló definiálása multi-class klasszifikációhoz.

```python

class ImageMultiClassClassificationNet(nn.Module):

    def __init__(self, n_categories):
        super().__init__()
        self.conv_block1 = nn.Sequential(
                # TODO start with this one: nn.Conv2d(1, 16, kernel_size=(3,3), padding='same')
                nn.Conv2d(n_categories, 16, kernel_size=(3,3), padding='same'),   # (bs,3,240,320) -> (bs,16,240,320)
                nn.ReLU(),
                nn.MaxPool2d(kernel_size=(2,2))   # (bs,16,240,320) -> (bs,16,120,160)
        )
        self.conv_block2 = nn.Sequential(
                nn.Conv2d(16, 32, kernel_size=(3,3), padding='same'),   # (bs,16,120,160) -> (bs,32,120,160)
                nn.ReLU(),
                nn.MaxPool2d(kernel_size=(2,2))   # (bs,16,120,160) -> (bs,32,60,80)
        )
        self.conv_block3 = nn.Sequential(
                nn.Conv2d(32, 32, kernel_size=(3,3), padding='same'),   # (bs,32,60,80) -> (bs,32,60,80)
                nn.ReLU(),
                nn.MaxPool2d(kernel_size=(2,2))   # (bs,32,60,80) -> (bs,32,30,40)
        )
        self.conv_block4 = nn.Sequential(
                nn.Conv2d(32, 32, kernel_size=(3,3), padding='same'),   # (bs,32,30,40) -> (bs,32,30,40)
                nn.ReLU(),
                nn.MaxPool2d(kernel_size=(2,2))   # (bs,32,30,40) -> (bs,32,15,20)
        )
        self.conv_block5 = nn.Sequential(
                nn.Conv2d(32, 16, kernel_size=(3,3), padding='same'),   # (bs,32,15,20) -> (bs,16,15,20)
                nn.ReLU(),
                nn.MaxPool2d(kernel_size=(2,2))   # (bs,16,15,20) -> (bs,16,7,10)
        )
        self.fcs = nn.Sequential(
                nn.Linear(16*7*10, 30),           # (bs,16*7*10) -> (bs,30)
                nn.ReLU(),
                nn.Linear(30, n_categories)       # (bs,30) -> (bs, n_categories)
        )

    def forward(self, x):
        # x.shape == (batch_size, 3, 240, 320)
        out = self.conv_block1(x)
        out = self.conv_block2(out)
        out = self.conv_block3(out)
        out = self.conv_block4(out)
        out = self.conv_block5(out)
        out = out.reshape(out.size(0), -1)
        out = self.fcs(out)
        return out

# instantiate net class

cl_model = ImageMultiClassClassificationNet(n_categories=3)
```

## Automatikus adat-augmentáció a Torchvision csomag segítségével az adat-iterátorban

Az adat-iterátor elkészítése (`torch.utils.data.Dataset` és `torch.utils.data.DataLoader` segítségével, `torchvision.transforms` által megvalósított adat-augmentációval).

```python
# very simple iterator: only basic type/dtype conversions happen inside + torchvision transforms calls

class ImageClassificationDataset_WithDataAugmentation(Dataset):
    def __init__(self, color_ims_split, cl_labels_split, transform):
        # expecting: color_ims_split (n_img, 240, 320, 3), cl_labels_split (n_img,)
        assert color_ims_split.shape[0] == cl_labels_split.shape[0]
        assert color_ims_split.shape[1:] == (240, 320, 3)
        self.color_ims_split = color_ims_split
        self.transform = transform

        #color_ims_split = color_ims_split.transpose(0, 3, 1, 2)       # LINE REMOVED; no need to do it here: transforms.v2.ToImage() does this now instead

        #self.xs_split_t = torch.from_numpy(color_ims_split)       # LINE REMOVED; no need to do it here: transforms.v2.ToImage() does this now instead
        self.ys_split_t = torch.from_numpy(cl_labels_split)       # tensor(n_img,) int64; multi-class labels

    def __len__(self):
        return self.color_ims_split.shape[0]

    def __getitem__(self, idx):
        # x = self.xs_split_t[idx,:,:,:].to(torch.float32)             # LINE REMOVED; no need to do it here: transforms.v2.ToDtype() does this now instead
        x_npy = self.color_ims_split[idx,:,:,:]
        y = self.ys_split_t[idx]
        x = self.transform(x_npy)           # LINE ADDED; ndarray(h, w, 3) uint8 range in [0, 255] -> Tensor(3, h, w) fl32 range in [0, 1]

        return x, y                # x: tensor(3, h, w) fl32 in range [0, 1];    y: scalar int64
                                   #     -> DataLoader will produce (tensor(batch_size, 3, h, w) fl32, tensor(batch_size,) int64)



train_transform = transforms.v2.Compose([
    transforms.v2.ToImage(),                             # ndarray(H, W, C) uint8 range in [0, 255] -> Tensor(C, H, W) uint8 range in [0, 255]
    transforms.v2.ToDtype(torch.float32, scale=True),   # Tensor(C, H, W) uint8 range in [0, 255] -> Tensor(C, H, W) fl32 range in [0, 1] as 'scale=True'
    transforms.v2.RandomHorizontalFlip(),
    transforms.v2.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2),
    transforms.v2.RandomAffine(degrees=15., translate=(0.2, 0.2), scale=(0.8, 1.2))
])

eval_transform = transforms.v2.Compose([
    transforms.v2.ToImage(),                             # ndarray(H, W, C) uint8 range in [0, 255] -> Tensor(C, H, W) uint8 range in [0, 255]
    transforms.v2.ToDtype(torch.float32, scale=True)    # Tensor(C, H, W) uint8 range in [0, 255] -> Tensor(C, H, W) fl32 range in [0, 1] as 'scale=True'
])



train_dataset = ImageClassificationDataset_WithDataAugmentation(color_ims_train, cl_labels_train, transform=train_transform)   # data-augmentation only here
val_dataset = ImageClassificationDataset_WithDataAugmentation(color_ims_val, cl_labels_val, transform=eval_transform)          # no need to augment during validation
test_dataset = ImageClassificationDataset_WithDataAugmentation(color_ims_test, cl_labels_test, transform=eval_transform)       # no need to augment during test

dataloader_cl_train = DataLoader(train_dataset, batch_size=32, shuffle=True)
dataloader_cl_val = DataLoader(val_dataset, batch_size=32, shuffle=True)
dataloader_cl_test = DataLoader(test_dataset, batch_size=32, shuffle=False)  # not shuffling!
```

```python
# implement your solution BELOW

@torch.no_grad()
def acc_fn(y_hat, y_true):
  # y_hat: (batch_size, n_cat) logits, y_true: (batch_size,) int64
    y_hat_am = torch.argmax(y_hat, dim=1)
    return torch.count_nonzero(y_hat_am == y_true) / y_hat.shape[0]

loss_fn = nn.CrossEntropyLoss()

n_epochs = 80
learning_rate = 0.001
optimizer = torch.optim.Adam(cl_model.parameters(), lr=learning_rate)

train_losses = []
val_losses = []
train_accuracy = []
val_accuracy = []
best_cl_model = None
best_val_loss = float('inf')
early_stopping_patience = 20
last_early_stopping_update_ep_idx = 0

for epoch_idx in range(n_epochs):
  cl_model.train()
  train_epoch_losses = []
  train_epoch_accs = []

  for x, y in dataloader_cl_train:
    x = x.to(TARGET_DEVICE)          # LINE CHANGED; 'tensor.to(TARGET_DEVICE)' could also be moved into the data iterators
    y = y.to(TARGET_DEVICE)          # LINE CHANGED

    optimizer.zero_grad()
    y_hat = cl_model(x)
    loss = loss_fn(y_hat, y)
    acc = acc_fn(y_hat, y)
    train_epoch_losses.append(loss.detach().to('cpu'))         # LINE CHANGED
    train_epoch_accs.append(acc.to('cpu'))                     # LINE CHANGED
    loss.backward()
    optimizer.step()

  # typical order of ops when moving data to TARGET_DEVICE for training/eval:
  #                                         torch.from_numpy(ndarray) -> tensor.to(TARGET_DEVICE)
  # typical order of ops when moving data on TARGET_DEVICE back to cpu:
  #                                         tensor.to('cpu') -> tensor.numpy()
  # typical order of ops when moving data connected to computational graph on TARGET_DEVICE back to cpu:
  #                                         tensor.detach() -> tensor.to('cpu') -> tensor.numpy()
  #                                         tensor.to('cpu') -> tensor.detach() -> tensor.numpy()    # also OK, no big difference in performance to previous case

  train_epoch_losses = torch.tensor(train_epoch_losses)
  train_epoch_accs = torch.tensor(train_epoch_accs)
  avg_tr_epoch_loss = train_epoch_losses.mean()
  avg_tr_epoch_acc = train_epoch_accs.mean()
  train_losses.append(avg_tr_epoch_loss)
  train_accuracy.append(avg_tr_epoch_acc)

  cl_model.eval()
  val_epoch_losses = []
  val_epoch_accs = []

  for x, y in dataloader_cl_val:
    x = x.to(TARGET_DEVICE)         # LINE CHANGED
    y = y.to(TARGET_DEVICE)         # LINE CHANGED

    with torch.no_grad():
      y_hat = cl_model(x)

    loss = loss_fn(y_hat, y)
    acc = acc_fn(y_hat, y)
    val_epoch_losses.append(loss.to('cpu'))         # LINE CHANGED
    val_epoch_accs.append(acc.to('cpu'))         # LINE CHANGED

  val_epoch_losses = torch.tensor(val_epoch_losses)
  val_epoch_accs = torch.tensor(val_epoch_accs)
  avg_val_epoch_loss = val_epoch_losses.mean()
  avg_val_epoch_acc = val_epoch_accs.mean()
  val_losses.append(avg_val_epoch_loss)
  val_accuracy.append(avg_val_epoch_acc)

  print(f"Epoch #{epoch_idx}, tr loss: {avg_tr_epoch_loss}, val loss: {avg_val_epoch_loss}, tr acc: {avg_tr_epoch_acc}, val acc: {avg_val_epoch_acc}")

  # early stopping: check if current weights give the lowest ever loss
  if avg_val_epoch_loss < best_val_loss:
    best_val_loss = avg_val_epoch_loss
    best_cl_model = copy.deepcopy(cl_model).to('cpu')       # LINE CHANGED; early stopping: stored best weights can also be moved off from TARGET_DEVICE to save up (possibly GPU) memory
    last_early_stopping_update_ep_idx = epoch_idx
    print("    Early stopping: Best model updated.")

  # early stopping: break if no update for a long time
  if last_early_stopping_update_ep_idx + early_stopping_patience < epoch_idx:
    print("Training is terminated by early stopping.")
    break

# early stopping: restore best weights
print("Best model weights are restored.")
cl_model = best_cl_model.to(TARGET_DEVICE)                  # LINE CHANGED; early stopping: finally, moving back stored weights to TARGET_DEVICE to overwrite current weights

# predict a few test samples (taking a single batch out of the data iterator and predicting labels for them
test_pred_y_cat = []
test_y_cat = []

with torch.no_grad():
    _, (test_x, test_y) = next(enumerate(dataloader_cl_test))
    test_x = test_x.to(TARGET_DEVICE)                                        # LINE CHANGED
    #test_pred_y = torch.nn.Softmax(dim=1)(cl_model(test_x))                 # if we need probabilities, a Softmax must be explicitly used on the network predicitons
    test_pred_y_cat_batch_npy = torch.argmax(cl_model(test_x), dim=1).to('cpu').numpy()         # LINE CHANGED
    test_pred_y_cat.append(test_pred_y_cat_batch_npy)
    test_y_cat.append(test_y.numpy())
test_pred_y_cat = np.concatenate(test_pred_y_cat, axis=0)
test_y_cat = np.concatenate(test_y_cat, axis=0)

# evaluating on test set
# Note: in case of images, the whole dataset is too large to be predicted at once, this would result in "RuntimeError: CUDA out of memory" on a GPU
#         therefore, test prediction must be batched too

with torch.no_grad():
  # compute test loss, accuracy
  test_pred_ys = []
  test_ys = []
  for x, y in dataloader_cl_test:
    y_hat = cl_model(x.to(TARGET_DEVICE)).to('cpu')         # LINE CHANGED
    test_pred_ys.append(y_hat)
    test_ys.append(y)

  test_pred_ys = torch.cat(test_pred_ys, dim=0)
  test_ys = torch.cat(test_ys, dim=0)
  test_ce = loss_fn(test_pred_ys, test_ys)
  test_acc = acc_fn(test_pred_ys, test_ys)
  print(f"Test metrics - CE: {test_ce}, acc: {test_acc}")

for sample_idx in range(10):
  print(f"Test sample #{sample_idx}, pred vs true label: {test_pred_y_cat[sample_idx]} vs {test_y_cat[sample_idx]}")

plt.plot(train_losses, label='Training')
plt.plot(val_losses, label='Validation')
plt.plot(train_accuracy, label='Training acc')
plt.plot(val_accuracy, label='Validation acc')
plt.xlabel('Epochs')
plt.ylabel('CE loss / Accuracy')
plt.legend()
plt.show()

# konf matrix
test_pred_ys_am = torch.argmax(test_pred_ys, dim=1)
conf_mat = np.zeros((3, 3), dtype=np.int32)
np.add.at(conf_mat, (test_pred_ys_am, test_ys), 1)

fig, ax = plt.subplots(figsize=(4,4))
ax.matshow(conf_mat, cmap=plt.cm.Blues, alpha=0.3)
for i in range(conf_mat.shape[0]):
    for j in range(conf_mat.shape[1]):
        ax.text(x=j, y=i, s=conf_mat[i, j], va='center', ha='center', size='xx-large')

plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show();
```

## **Előtanított mély konvolúciós háló finomhangolása** képek (multi-class) klasszifikációjára, adat-augmentációval. (CPU vagy GPU.)

Az adat-iterátor elkészítése (`torch.utils.data.Dataset` és `torch.utils.data.DataLoader` segítségével, `torchvision.transforms` által megvalósított adat-augmentációval).

A MobileNet v2 háló speciális értéktartományokban várja az input adatokat, **ezért változtatni kell a korábbi adat-iterátor kódján**. Részletek: https://pytorch.org/hub/pytorch_vision_mobilenet_v2/

```python
# very simple iterator: only basic type/dtype conversions happen inside + torchvision transforms calls (including MobileNet v2 normalization)

MOBILENET_MEANS = [0.485, 0.456, 0.406]                  # LINE ADDED
MOBILENET_STDS = [0.229, 0.224, 0.225]                   # LINE ADDED; MobileNet v2 stats used for image normalization

class ImageClassificationDataset_WithDataAugmentation_MobileNet2(Dataset):
    def __init__(self, color_ims_split, cl_labels_split, transform):
        # expecting: color_ims_split (n_img, 240, 320, 3), cl_labels_split (n_img,)
        assert color_ims_split.shape[0] == cl_labels_split.shape[0]
        assert color_ims_split.shape[1:] == (240, 320, 3)
        self.color_ims_split = color_ims_split
        self.transform = transform

        #color_ims_split = color_ims_split.transpose(0, 3, 1, 2)       # no need to do it here: transforms.v2.ToImage() does this now instead

        #self.xs_split_t = torch.from_numpy(color_ims_split)       # no need to do it here: transforms.v2.ToImage() does this now instead
        self.ys_split_t = torch.from_numpy(cl_labels_split)       # tensor(n_img,) int64; multi-class labels

    def __len__(self):
        return self.color_ims_split.shape[0]

    def __getitem__(self, idx):
        # x = self.xs_split_t[idx,:,:,:].to(torch.float32)             # no need to do it here: transforms.v2.ToDtype() does this now instead
        x_npy = self.color_ims_split[idx,:,:,:]
        y = self.ys_split_t[idx]
        x = self.transform(x_npy)           # ndarray(h, w, 3) uint8 range in [0, 255] -> Tensor(3, h, w) fl32 range in [0, 1]

        return x, y                # x: tensor(3, h, w) fl32 in range [0, 1];    y: scalar int64
                                   #     -> DataLoader will produce (tensor(batch_size, 3, h, w) fl32, tensor(batch_size,) int64)


train_transform = transforms.v2.Compose([
    transforms.v2.ToImage(),                             # ndarray(H, W, C) uint8 range in [0, 255] -> Tensor(C, H, W) uint8 range in [0, 255]
    transforms.v2.ToDtype(torch.float32, scale=True),   # Tensor(C, H, W) uint8 range in [0, 255] -> Tensor(C, H, W) fl32 range in [0, 1] as 'scale=True'
    transforms.v2.RandomHorizontalFlip(),
    transforms.v2.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2),
    transforms.v2.RandomAffine(degrees=15., translate=(0.2, 0.2), scale=(0.8, 1.2)),
    transforms.v2.Normalize(mean=MOBILENET_MEANS, std=MOBILENET_STDS)                 # LINE ADDED
])

eval_transform = transforms.v2.Compose([
    transforms.v2.ToImage(),                             # ndarray(H, W, C) uint8 range in [0, 255] -> Tensor(C, H, W) uint8 range in [0, 255]
    transforms.v2.ToDtype(torch.float32, scale=True),    # Tensor(C, H, W) uint8 range in [0, 255] -> Tensor(C, H, W) fl32 range in [0, 1] as 'scale=True'
    transforms.v2.Normalize(mean=MOBILENET_MEANS, std=MOBILENET_STDS)                 # LINE ADDED
])



train_dataset = ImageClassificationDataset_WithDataAugmentation_MobileNet2(color_ims_train, cl_labels_train, transform=train_transform)   # data-augmentation only here
val_dataset = ImageClassificationDataset_WithDataAugmentation_MobileNet2(color_ims_val, cl_labels_val, transform=eval_transform)          # no need to augment during validation
test_dataset = ImageClassificationDataset_WithDataAugmentation_MobileNet2(color_ims_test, cl_labels_test, transform=eval_transform)       # no need to augment during test

dataloader_cl_train = DataLoader(train_dataset, batch_size=32, shuffle=True)
dataloader_cl_val = DataLoader(val_dataset, batch_size=32, shuffle=True)
dataloader_cl_test = DataLoader(test_dataset, batch_size=32, shuffle=False)  # not shuffling!
```

1. verzió: MobileNet v2 + 1 x konvolúciós réteg + 2 x teljesen összekötött réteg

```python
class MobileNet2_FineTunedForMultiClassClassification(nn.Module):

    def __init__(self, n_categories):
        super().__init__()
        self.mobile_net = torch.hub.load('pytorch/vision', 'mobilenet_v2', weights='MobileNet_V2_Weights.IMAGENET1K_V1')
        self.mobile_net_output = None
        self.mobile_net.features[17].conv[3].register_forward_hook(self._hook)
        for param in self.mobile_net.parameters():            # freeze all weights of the original MobileNet network
          param.requires_grad = False                         #      good if we have very little training data and our task is related to an image classification task on ImageNet
                                                              #      bad if our task is highly different from an image classification task on ImageNet

        self.extra_conv = nn.Conv2d(320, 40, kernel_size = (3,3), padding='same')  # out: (bs, 40, 8, 10) if input size is (240, 320)
        self.extra_fc = nn.Sequential(
            nn.ReLU(),
            nn.Linear(40*8*10, 20),
            nn.ReLU(),
            nn.Linear(20, n_categories)          # multi-class classification: no activation function in PyTorch at end
        )

    def _hook(self, module, input, output):
        self.mobile_net_output = output   # (batch_size, 320, 4, 4)

    def forward(self, x):
        _ = self.mobile_net(x)
        #print("SH", self.mobile_net_output.shape)   # -> SH torch.Size([32, 320, 8, 10]) if input size is (240, 320)
        out = self.extra_conv(self.mobile_net_output)
        out = out.reshape(out.size(0), -1)
        out = self.extra_fc(out)
        return out


# instantiate net class

cl_model = MobileNet2_FineTunedForMultiClassClassification(n_categories=3)

print(cl_model)
```

```python
# implement your solution BELOW

@torch.no_grad()
def acc_fn(y_hat, y_true):
  # y_hat: (batch_size, n_cat) logits, y_true: (batch_size,) int64
    y_hat_am = torch.argmax(y_hat, dim=1)
    return torch.count_nonzero(y_hat_am == y_true) / y_hat.shape[0]

loss_fn = nn.CrossEntropyLoss()

n_epochs = 20
learning_rate = 0.001
optimizer = torch.optim.Adam(cl_model.parameters(), lr=learning_rate)

train_losses = []
val_losses = []
train_accuracy = []
val_accuracy = []
best_cl_model = None
best_val_loss = float('inf')
early_stopping_patience = 5
last_early_stopping_update_ep_idx = 0

for epoch_idx in range(n_epochs):
  cl_model.train()
  train_epoch_losses = []
  train_epoch_accs = []

  for x, y in dataloader_cl_train:
    x = x.to(TARGET_DEVICE)          # LINE CHANGED; 'tensor.to(TARGET_DEVICE)' could also be moved into the data iterators
    y = y.to(TARGET_DEVICE)          # LINE CHANGED

    optimizer.zero_grad()
    y_hat = cl_model(x)
    loss = loss_fn(y_hat, y)
    acc = acc_fn(y_hat, y)
    train_epoch_losses.append(loss.detach().to('cpu'))         # LINE CHANGED
    train_epoch_accs.append(acc.to('cpu'))                     # LINE CHANGED
    loss.backward()
    optimizer.step()

  # typical order of ops when moving data to TARGET_DEVICE for training/eval:
  #                                         torch.from_numpy(ndarray) -> tensor.to(TARGET_DEVICE)
  # typical order of ops when moving data on TARGET_DEVICE back to cpu:
  #                                         tensor.to('cpu') -> tensor.numpy()
  # typical order of ops when moving data connected to computational graph on TARGET_DEVICE back to cpu:
  #                                         tensor.detach() -> tensor.to('cpu') -> tensor.numpy()
  #                                         tensor.to('cpu') -> tensor.detach() -> tensor.numpy()    # also OK, no big difference in performance to previous case

  train_epoch_losses = torch.tensor(train_epoch_losses)
  train_epoch_accs = torch.tensor(train_epoch_accs)
  avg_tr_epoch_loss = train_epoch_losses.mean()
  avg_tr_epoch_acc = train_epoch_accs.mean()
  train_losses.append(avg_tr_epoch_loss)
  train_accuracy.append(avg_tr_epoch_acc)

  cl_model.eval()
  val_epoch_losses = []
  val_epoch_accs = []

  for x, y in dataloader_cl_val:
    x = x.to(TARGET_DEVICE)         # LINE CHANGED
    y = y.to(TARGET_DEVICE)         # LINE CHANGED

    with torch.no_grad():
      y_hat = cl_model(x)

    loss = loss_fn(y_hat, y)
    acc = acc_fn(y_hat, y)
    val_epoch_losses.append(loss.to('cpu'))         # LINE CHANGED
    val_epoch_accs.append(acc.to('cpu'))         # LINE CHANGED

  val_epoch_losses = torch.tensor(val_epoch_losses)
  val_epoch_accs = torch.tensor(val_epoch_accs)
  avg_val_epoch_loss = val_epoch_losses.mean()
  avg_val_epoch_acc = val_epoch_accs.mean()
  val_losses.append(avg_val_epoch_loss)
  val_accuracy.append(avg_val_epoch_acc)

  print(f"Epoch #{epoch_idx}, tr loss: {avg_tr_epoch_loss}, val loss: {avg_val_epoch_loss}, tr acc: {avg_tr_epoch_acc}, val acc: {avg_val_epoch_acc}")

  # early stopping: check if current weights give the lowest ever loss
  if avg_val_epoch_loss < best_val_loss:
    best_val_loss = avg_val_epoch_loss
    best_cl_model = copy.deepcopy(cl_model).to('cpu')       # LINE CHANGED; early stopping: stored best weights can also be moved off from TARGET_DEVICE to save up (possibly GPU) memory
    last_early_stopping_update_ep_idx = epoch_idx
    print("    Early stopping: Best model updated.")

  # early stopping: break if no update for a long time
  if last_early_stopping_update_ep_idx + early_stopping_patience < epoch_idx:
    print("Training is terminated by early stopping.")
    break

# early stopping: restore best weights
print("Best model weights are restored.")
cl_model = best_cl_model.to(TARGET_DEVICE)                  # LINE CHANGED; early stopping: finally, moving back stored weights to TARGET_DEVICE to overwrite current weights

# predict a few test samples (taking a single batch out of the data iterator and predicting labels for them
test_pred_y_cat = []
test_y_cat = []

with torch.no_grad():
    _, (test_x, test_y) = next(enumerate(dataloader_cl_test))
    test_x = test_x.to(TARGET_DEVICE)                                        # LINE CHANGED
    #test_pred_y = torch.nn.Softmax(dim=1)(cl_model(test_x))                 # if we need probabilities, a Softmax must be explicitly used on the network predicitons
    test_pred_y_cat_batch_npy = torch.argmax(cl_model(test_x), dim=1).to('cpu').numpy()         # LINE CHANGED
    test_pred_y_cat.append(test_pred_y_cat_batch_npy)
    test_y_cat.append(test_y.numpy())
test_pred_y_cat = np.concatenate(test_pred_y_cat, axis=0)
test_y_cat = np.concatenate(test_y_cat, axis=0)

# evaluating on test set
# Note: in case of images, the whole dataset is too large to be predicted at once, this would result in "RuntimeError: CUDA out of memory" on a GPU
#         therefore, test prediction must be batched too

with torch.no_grad():
  # compute test loss, accuracy
  test_pred_ys = []
  test_ys = []
  for x, y in dataloader_cl_test:
    y_hat = cl_model(x.to(TARGET_DEVICE)).to('cpu')         # LINE CHANGED
    test_pred_ys.append(y_hat)
    test_ys.append(y)

  test_pred_ys = torch.cat(test_pred_ys, dim=0)
  test_ys = torch.cat(test_ys, dim=0)
  test_ce = loss_fn(test_pred_ys, test_ys)
  test_acc = acc_fn(test_pred_ys, test_ys)
  print(f"Test metrics - CE: {test_ce}, acc: {test_acc}")

for sample_idx in range(10):
  print(f"Test sample #{sample_idx}, pred vs true label: {test_pred_y_cat[sample_idx]} vs {test_y_cat[sample_idx]}")

plt.plot(train_losses, label='Training')
plt.plot(val_losses, label='Validation')
plt.plot(train_accuracy, label='Training acc')
plt.plot(val_accuracy, label='Validation acc')
plt.xlabel('Epochs')
plt.ylabel('CE loss / Accuracy')
plt.legend()
plt.show()

# konf matrix
test_pred_ys_am = torch.argmax(test_pred_ys, dim=1)
conf_mat = np.zeros((3, 3), dtype=np.int32)
np.add.at(conf_mat, (test_pred_ys_am, test_ys), 1)

fig, ax = plt.subplots(figsize=(4,4))
ax.matshow(conf_mat, cmap=plt.cm.Blues, alpha=0.3)
for i in range(conf_mat.shape[0]):
    for j in range(conf_mat.shape[1]):
        ax.text(x=j, y=i, s=conf_mat[i, j], va='center', ha='center', size='xx-large')

plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show();
```
